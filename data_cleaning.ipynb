```python
# creating a new column performance status
ai['Performance_Status'] = ai.apply(lambda x: 'Improved' if x['Post_Semester_GPA'] > x['Pre_Semester_GPA'] else ('Declined' if x['Post_Semester_GPA'] < x['Pre_Semester_GPA'] else 'No Change'), axis=1)
```

```python
ai.head(10)
```

```python
#creating a new column
labels = ['Low' , 'Medium' , 'High']
ai['ai_dependency_level'] = pd.qcut(ai['Perceived_AI_Dependency'], q=3 , labels = labels) 
ai['Anxiety_Level'] = pd.qcut(ai['Anxiety_Level_During_Exams'], q=3 , labels = labels) 
```

```python
ai.columns
ai.head(10)
```

```python
# converting column names in proper format
ai.columns = ai.columns.str.lower()
ai.columns = ai.columns.str.replace(' ','_')
```
